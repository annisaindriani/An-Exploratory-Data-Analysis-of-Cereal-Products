**Extracting and replacing values**

```python
manufacturers = {
    'A' : 'American Home Food Products',
    'G' : 'General Mills',
    'K' : 'Kelloggs',
    'N' : 'Nabisco',
    'P' : 'Post',
    'Q' : 'Quaker Oats',
    'R' : 'Ralston Purina',
 }
served = {
    'C': 'cold',
    'H': 'hot',
}
nutrition_list = [
    'calories',
    'protein',
    'fat',
    'sodium',
    'fiber',
    'carbohydrates',
    'sugars',
    'potassium',
    'vitamins'
]
```

```python
def change_columns_values(df,list_items,column_name):
    for index, value in list_items.items():
        df.loc[df[column_name] == index, column_name] = value

    return df
#Change columns name into understandable text
df = df.rename(columns={"mfr": "manufacturer", "type": "served", "potass":"potassium", "carbo": "carbohydrates"})
df = change_columns_values(df,manufacturers,'manufacturer')
df = change_columns_values(df,served,'served')
for nl in nutrition_list:
    df.loc[df[nl] == -1, nl] = 0 #In reality, you can't have -1 values on the nutrition columns
```

```python
df
```

Before the dataset is full of abbreviation, which seems hard to understand what the data means. Hence, after replacing those abbreviation, we can see name of the manufacturer and what it is usually served (cold or hot). 

**Convert the data from one format to another** <br>
Convert 'served' feature into numeric

```python
serving_mapping ={'cold': 1, 'hot':0}
data = [df]
for dataset in data:
    dataset['served'] = dataset['served'].map(serving_mapping)
    dataset['served'] = dataset['served'].fillna(0)
```

Convert 'fiber', 'carbohydrates', and 'weight' from float to int using astype.

```python
data = [df]
for dataset in data:
    dataset['fiber'] = dataset['fiber'].fillna(0)
    dataset['fiber'] = dataset['fiber'].astype(int)
    dataset['carbohydrates'] = dataset['carbohydrates'].fillna(0)
    dataset['carbohydrates'] = dataset['carbohydrates'].astype(int)
    dataset['weight'] = dataset['weight'].fillna(0)
    dataset['weight'] = dataset['weight'].astype(int)
```

```python
df
```

Because there's only two options, cold or hot so we decide to change it like boolean value (1 or 0). Nutrition list (protein, fat, etc) is mostly in integer form, so we change the others that is in float form into integer.

We can see that the cereal that has the highest rating has the lowest calories. The higher the calories contained in the cereal, the lower the rating tends to be. It can be concluded that people prefer cereals that are low in calories.

**Test significant variables(correlation)** <br>
Heatmap of Correlation between different features: <br>
Positive numbers = Positive correlation, i.e. increase in one feature will increase the other feature & vice-versa. <br>
Negative numbers = Negative correlation, i.e. increase in one feature will decrease the other feature & vice-versa. <br>

In our case, we focus on which features have strong positive or negative correlation with the calories feature.

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.drop('calories', axis=1).corr(), vmax=0.6, square=True, annot=True)
```

The highest positive correlation remains with itself. But among all, the one with the high positive correlation is between fiber and potassium. There are a lot of possibilities  but they stay aligned. Meanwhile, there is a negative correlation between rating and sugar, which means that sugar will not affect the rating in any way.
